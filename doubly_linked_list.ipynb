```python
#Circularly doubly linked list
class Node:
    def __init__(self, key = None):
        self.key = key
        self.next = self
        self.prev = self
            
    def __str__(self):
        return str(self.key)
```

```python
class CurcularlyDoublyLinkedList:
    def __init__(self):
        self.head = Node()
        self.size = 0
        
    def __len__(self):
        v = self.head
        while v.next != self.head:
            v = v.next
            self.size +=1

        return self.size
    
    def splice(self, a, b, x): #O(1)
        ap = a.prev
        bn = b.next
        xn = x.next

        ap.next = bn
        bn.prev = ap

        x.next = a
        a.prev = x
        b.next = xn
        xn.prev = b

    def moveAfter(self, a, x):
        return splice(a, a, x)
    
    def moveBefore(self, a, x):
        return splice(a, a, x.prev)
    
    def insertAfter(self, x, key):
        new_node = Node(key)
        return moveAfter(new_node, x)
    
    def insertBefore(self, x, key):
        new_node = Node(key)
        return moveBefore(new_node, x)
    
    def pushFront(self, key):
        new_node = Node(key)
        return insertBefore(new_node, self.head)
    
    def pushBack(self, key):
        new_node = Node(key)
        return insertBefore(self.head, new_node)

    def search(self, key): #O(n)
        v = self.head #from dummy node
        while v.next != self.head:
            if v.key == key:
                return v
            v = v.next
        return None
    
    def remove(self, x):
        xp = x.prev
        xn = x.next
        if x == None or x == self.head:
            return
        else:
            xp.next = xn
            xn.prev = xp

            del x

    def popFront(self):
        if self.size == 0 :
            return None
        else:
            return remove(self.head.next)
    
    def popBack(self):
        if self.size == 0:
            return None
        else:
            return remove(self.head.prev)
        
    def popNode(self, key):
        obj_node = search(key)
        return remove(obj_node)

        
```
