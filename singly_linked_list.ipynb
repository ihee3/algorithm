```python
class Node:
    def __init__(self, key = None):
        self.key = key
        self.next_node = None

    def __str__(self):
        return str(self.key)
```

```python
a = Node(3)
b = Node(9)
c = Node(-1)

a.next_node = b
b.next_node = c #3 -> 9 -> -1 -> None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    def pushFront(self, key): # O(1)
        new_node= Node(key)
        new_node.next_node = self.head
        self.head = new_node
        self.size += 1
    
    def pushBack(self, key): # O(n)
        new_node = Node(key)
        if len(self) == 0:
            self.head = new_node
        else:
            tail = self.head
            while tail.next_node != None:
                tail = tail.next
            tail.next = new_node
        
        self.size += 1
    
    def __len__(self):
        return self.size
    
    def popFront(self): # O(1)
        if len(self) ==0:
            return None
        else:
            x = self.head
            key = x.key
            self.head = x.next_node
            self.size -= 1
            del x
            return key

    def popBack(self): # O(n)
        if len(self) == 0:
            return None
        else:
            prev, tail = None, self.head
            while tail.next_node != None :
                prev = tail
                tail = tail.next_node
            
            if len(self) == 1:
                self.head = None
            else:
                prev.next_node = tail.next_node

            key = tail.key
            del tail
            self.size -= 1 

            return key
        
    def search(self, key): #O(n)
            v = self.head
            while v != None :
                if v.key == key:
                    return v
                v = v.next_node
            return None
        
    def __iterator__(self): #linked list를 for 문을 사용할 수 있도록 하는 generator * yield *
            v = self.head
            while v != None:
                yield v #return과 비슷한 역할 
                v = v.next_node
                return
            
```
